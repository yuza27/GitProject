```python
raw_data = pd.read_csv('Customer_Data.csv')
raw_data
```

```python
X = raw_data.drop(['Customer_ID','CustCreationDate','Date_Of_Last_ Purchase','Unnamed: 11','Unnamed: 12'],axis=1)
X
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
X_pca
```
